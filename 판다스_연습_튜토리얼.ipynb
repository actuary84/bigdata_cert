#groupby 학습



````python
import seaborn as sns
df1 = sns.load_dataset('penguins')
df1
````

````python
#species에 따라 그룹으로 나눔

df_split = df1.groupby('species')
df_split.mean()
````

````python
#기준을 여러개로 나눔

df_split2 = df1.groupby(['species','sex'])
df_split2.mean()
````

````python
#사용자 함수를 직접 만든 후 그룹 객체에 적용할 때는 agg()메서드 사용
def max_min(x):
  return x.max() - x.min()
````

````python
#species별로 그룹을 나눈 후, bill_length_mm 열만 선택하여 min_max 함수 적용
df1.groupby('species')['bill_length_mm'].agg(max_min)
````

````python
#리스트의 형태로 입력하면 동시에 함수 적용할 수 있음
df1.groupby('species').agg(['max','min'])
````

````python
#딕셔너리 형태로 각 열마다 다른 종류의 함수를 적용할 수 있음
df1.groupby('species').agg({'bill_length_mm':['min','max'],'island':'count'})
````

````python
df1.groupby('species')['bill_length_mm'].agg('mean')
````

````python
#transform() 메서드는 계산의 결과를 본래의 행 인덱스와 열 인덱스를 기준으로 반환하여 데이터프레임과 같은 형태로 변형해서 데이터를 정리해줌
df1.groupby('species')['bill_length_mm'].transform('mean')
````

````python
#각 그룹별 z-score(각 데이터가 평균으로부터 얼마나 떨어져 있는지)를 계산
def z_score(x):
  z = (x-x.mean())/x.std()
  return z

#그룹별로 z-score를 구한 후 원래 행 부분에 데이터를 넣어줌
df1.groupby('species')['bill_length_mm'].transform(z_score)
````

````python
#그룹 객체에도 apply 이용하여 적용가능
df1.groupby('species')['bill_length_mm'].apply('min')
````

````python
df1.groupby('species')['bill_length_mm'].transform(z_score)
````

````python
#그룹 객체에 filter() 메서드를 적용하여 bill_length_mm가 40이상인 종만 찾기

df1.groupby('species').filter(lambda x: x['bill_length_mm'].mean()>=40)
````

##concat()함수
- 행(axis=0, 디폴트) 또는 열 방향(axis=1)으로 이어 붙이는 개념, 열 방향으로 붙일 때는 행 인덱스를 기준, 디폴트로 합집합(join='outer')
- join='left', join='right'는 지원하지 않음 


````python
#예제 데이터 프레임 생성

df1 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
},index=[0, 1, 2, 3],
)

df2 = pd.DataFrame({
    "A": ["A4", "A5", "A6", "A7"],
    "B": ["B4", "B5", "B6", "B7"],
    "C": ["C4", "C5", "C6", "C7"],
    "D": ["D4", "D5", "D6", "D7"]
},
    index=[4, 5, 6, 7],
)

df3 = pd.DataFrame({
    "A": ["A8", "A9", "A10", "A11"],
    "B": ["B8", "B9", "B10", "B11"],
    "C": ["C8", "C9", "C10", "C11"],
    "D": ["D8", "D9", "D10", "D11"]
},
    index=[8, 9, 10, 11],
)

print('df1\n',df1)
print('df2\n',df2)
print('df3\n',df3)
````

````python
#행 방향으로 데이터 프레임 합치기

pd.concat([df1,df2,df3])
````

````python
#shape이 다른 데이터프레임 행방향으로 합치기

df4 = pd.DataFrame({
    "B": ["B2", "B3", "B6", "B7"],
    "D": ["D2", "D3", "D6", "D7"],
    "F": ["F2", "F3", "F6", "F7"]
},
    index=[2, 3, 6, 7]
)

print('df1\n',df1)
print('\ndf4\n',df4)
````

````python
pd.concat([df1,df4])
````

````python
#인덱스 초기화
pd.concat([df1,df4],ignore_index=True) 
````

````python
print('df1\n',df1)
print('\ndf4\n',df4)
````

````python
#열방향으로 데이터 프레임 합치기 -> 디폴트로 join='outer'
pd.concat([df1,df4],axis=1) 
````

````python
#열방향으로 동일한 인덱스만 합치기(행인 덱스가 기준) -> 공통적으로 행인덱스가 존재하는 애들만  출력됨
pd.concat([df1,df4],axis=1, join='inner')
````

##merge()함수
- 키를 기준으로 데이터프레임을 합침
- inner join, left join, right join, outer join 방법이 있음
- left join : 왼쪽은 모두 유지, 오른쪽이 키를 기준으로 합쳐짐
- right join : 오른쪽은 모두 유지, 왼쪽이 키를 기준으로 합쳐짐

````python
#예제 데이터프레임 생성
left = pd.DataFrame({
    "key": ["K0", "K1", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
})


right = pd.DataFrame({
    "key": ["K0", "K1", "K3", "K4"],
    "C": ["C0", "C1", "C3", "C4"],
    "D": ["D0", "D1", "D3", "D4"],
})

print('left\n',left)
print('\nright\n',right)
````

````python
#inner join -> 디폴트
pd.merge(left,right,on='key')
````

````python
#left join
pd.merge(left,right,on='key',how='left')
````

````python
#right join
pd.merge(left,right,on='key',how='right')
````

````python
#outer join
pd.merge(left,right,on='key',how='outer')
````

````python
#기준이 되는 키 열이 서로 다른 경우 on 대신 left_on과 right_on을 통해 키를 직접 선언

# 키 인덱스가 서로 다른 데이터프레임 생성
left = pd.DataFrame({
    "key_left": ["K0", "K1", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
})


right = pd.DataFrame({
    "key_right": ["K0", "K1", "K3", "K4"],
    "C": ["C0", "C1", "C3", "C4"],
    "D": ["D0", "D1", "D3", "D4"],
})

print('left\n',left)
print('\nright\n',right)
````

````python
#left_on과 right_on을 이용한 inner join

pd.merge(left,right,left_on='key_left', right_on='key_right', how='inner')
````

````python
#실무에서는 직관적인 표현으로 다음과 같이 사용한다

left.merge(right, left_on='key_left', right_on='key_right', how='inner')
````

##join()메서드
- merge()함수는 키를 기준으로 결합을, join()메서드는 행 인덱스를 기준으로 합친다

````python
#예제 데이터프레임 생성

left = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]},
    index=["K0", "K1", "K2", "K3"]
)


right = pd.DataFrame({
    "C": ["C0", "C1", "C3", "C4"],
    "D": ["D0", "D1", "D3", "D4"]},
    index=["K0", "K1", "K3", "K4"])

print('left\n',left)
print('\nright\n',right)
````

````python
#행 인덱스를 기준으로 합쳐짐
left.join(right)
````
